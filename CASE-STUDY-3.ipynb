```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('UDISE.CSV')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.fillna(df.median(numeric_only=True), inplace=True)
```

```python
df.info()
```

```python
numeric_cols = df.columns[1:]
df
```

```python
fig, axes = plt.subplots(5, 2, figsize=(16, 20))
fig.suptitle('Data Analysis Visualizations', fontsize=16)
```

```python
fig, axes = plt.subplots(5, 2, figsize=(16, 20))
fig.suptitle('Data Analysis Visualizations', fontsize=16)
```

```python
for i, col in enumerate(numeric_cols[:10]):
    row, col_idx = divmod(i, 2)
    sns.histplot(df[col], bins=15, kde=True, ax=axes[row, col_idx])
    axes[row, col_idx].set_title(f'Histogram of {col}', fontsize=12)
plt.tight_layout(rect=[0, 0.03, 1, 0.97])
plt.show()
```

```python
plt.figure(figsize=(16, 20))

# Line Plot
plt.subplot(3, 2, 1)
sns.lineplot(data=df[numeric_cols[:5]])
plt.title("Line Plot of First 5 Numeric Columns")
```

```python
plt.subplot(3, 2, 2)
sns.boxplot(data=df[numeric_cols[:5]])
plt.title("Box Plot of First 5 Numeric Columns")
```

```python
sns.pairplot(df[numeric_cols[:5]])
plt.suptitle("Pairplot of First 5 Numeric Columns", y=1.02)
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df[numeric_cols[0]], y=df[numeric_cols[1]])
plt.title(f"Scatter Plot: {numeric_cols[0]} vs {numeric_cols[1]}")
plt.xlabel(numeric_cols[0])
plt.ylabel(numeric_cols[1])
plt.show()
```

```python
plt.tight_layout(rect=[0, 0.03, 1, 0.97])
plt.show()
```

```python
plt.figure(figsize=(16, 30))
plt.subplot(5, 2, 1)
sns.lineplot(data=df[numeric_cols[:5]])
plt.title("Line Plot of First 5 Numeric Columns")
```

```python
sns.pairplot(df[numeric_cols[:5]])
plt.suptitle("Pairplot of First 5 Numeric Columns", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(16, 25))
plt.subplot(3, 2, 2)
sns.barplot(x=df[numeric_cols[0]], y=df[numeric_cols[1]])
plt.title(f"Bar Plot: {numeric_cols[0]} vs {numeric_cols[1]}")
plt.show()
```

```python
plt.figure(figsize=(16, 25))
plt.subplot(3, 2, 4)
sns.countplot(x=df[numeric_cols[0]])
plt.title(f"Count Plot of {numeric_cols[0]}")
```

```python
plt.figure(figsize=(16, 25))
plt.subplot(3, 2, 5)
sns.stripplot(data=df[numeric_cols[:5]], jitter=True)
plt.title("Strip Plot of First 5 Numeric Columns")

plt.tight_layout()
plt.show()
```
